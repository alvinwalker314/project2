# Data Wrangling

```python
col_names= {
    'Time Code': 'Year',
    'Access to Clean Fuels and Technologies for cooking (% of total population) [2.1_ACCESS.CFT.TOT]':'% of pop with access to clean fuels and technologies for cooking',
    'Access to electricity (% of rural population with access) [1.2_ACCESS.ELECTRICITY.RURAL]':'% rural pop with access to electricity',
    'Access to electricity (% of total population) [1.1_ACCESS.ELECTRICITY.TOT]': '% of pop with access to electricity',
    'Access to electricity (% of urban population with access) [1.3_ACCESS.ELECTRICITY.URBAN]':'% of urban pop with access to electricity',
    'Energy intensity level of primary energy (MJ/2011 USD PPP) [6.1_PRIMARY.ENERGY.INTENSITY]':'Energy intensity level of primary energy',
    'Renewable electricity output (GWh) [4.1.2_REN.ELECTRICITY.OUTPUT]':'Renewable electricity output (GWh)',
    'Renewable electricity share of total electricity output (%) [4.1_SHARE.RE.IN.ELECTRICITY]':'% of renewable electricty to total electricity output',
    'Renewable energy consumption (TJ) [3.1_RE.CONSUMPTION]': 'renewable energy consumption',
    'Total final energy consumption (TFEC) (TJ) [1.1_TOTAL.FINAL.ENERGY.CONSUM]':'total final energy consumption',
    'Total electricity output (GWh) [4.1.1_TOTAL.ELECTRICITY.OUTPUT]':'total electricity output (GWh)',
    'Renewable energy share of TFEC (%) [2.1_SHARE.TOTAL.RE.IN.TFEC]':'% of renewable energy to TFEC'}

data= data.rename(columns= col_names)
```

```python
columns= ['Country Name',
          'Year',
          '% of pop with access to clean fuels and technologies for cooking',
          '% rural pop with access to electricity',
          '% of pop with access to electricity',
          '% of urban pop with access to electricity',
          'Energy intensity level of primary energy',
#           'renewable energy consumption',
          'total final energy consumption',
          'Renewable electricity output (GWh)',
          'total electricity output (GWh)']

data= data[columns]
```

```python
data['Year']= data['Year'].str.strip('YR')

years= data['Year'].unique()

countries= data['Country Name'].unique()

data= data.replace('..', 0)
```

```python
columns= ['Year',
          '% of pop with access to clean fuels and technologies for cooking',
          '% rural pop with access to electricity',
          '% of pop with access to electricity',
          '% of urban pop with access to electricity',
          'Energy intensity level of primary energy',
          'Renewable electricity output (GWh)',
#           'renewable energy consumption',
          'total final energy consumption',
          'total electricity output (GWh)']

data[columns]= data[columns].astype(float)

data['Country Name']= data['Country Name'].astype(str)
```

```python
pipeline= make_pipeline(ce.OrdinalEncoder(), SimpleImputer(strategy= 'mean'))

data_encoded= pipeline.fit_transform(data)
```

```python
data_encoded= pd.DataFrame(data= data_encoded,columns=data.columns)

data_encoded= data_encoded.dropna()
```

# Train Test Validate

```python
train_years_0, test_years= train_test_split( years, train_size= 0.8, test_size= 0.2, random_state=42)
```

```python
train_years_0.shape, test_years.shape
```

```python
train_years, val_years= train_test_split(train_years_0, train_size=0.8, test_size= 0.2, random_state= 42)
```

```python
train_years.shape, val_years.shape, test_years.shape
```

```python
train= data_encoded[data_encoded['Year'].isin(train_years)]
val= data_encoded[data_encoded['Year'].isin(val_years)]
test= data_encoded[data_encoded['Year'].isin(test_years)]
```

```python
target= 'Renewable electricity output (GWh)'

train_target= train[target]
val_target= val[target]
test_target= test[target]

log_train_target= np.log1p(train_target)
log_val_target= np.log1p(val_target)
log_test_target= np.log1p(test_target)

train_features= train.drop(columns=target)
val_features= val.drop(columns=target)
test_features= test.drop(columns=target)
```

```python
target_mean= val_target.mean()
baseline= np.full_like(val_target, fill_value= target_mean)
log_baseline= np.full_like(baseline, fill_value= np.log1p(target_mean))
```

```python
val_target.shape, baseline.shape
```

```python
np.sqrt(mean_squared_error(baseline, val_target)), r2_score(baseline, val_target)
```

```python
np.sqrt(mean_squared_error(log_baseline, log_val_target)), r2_score(log_baseline, log_val_target)
```

```python
booster= XGBRegressor(n_estimators= 1000, random_state= 42, early_stopping= 10, n_jobs=-1)
booster.fit(train_features, train_target)
```

```python
val_pred= booster.predict(val_features)
test_pred= booster.predict(test_features)
```

```python
np.sqrt(mean_squared_error(val_pred, val_target)), r2_score(val_pred, val_target)
```

```python
np.sqrt(mean_squared_error(test_pred, test_target)), r2_score(test_pred, test_target)
```

```python
permuter = PermutationImportance(booster,cv='prefit', n_iter=2, random_state=42)

permuter.fit(val_features, log_val_target)
```

```python
feature_names = val_features.columns.tolist()

eli5.show_weights(permuter, top=None, feature_names=feature_names)
```
